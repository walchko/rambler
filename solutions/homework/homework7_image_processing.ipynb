```python
from __future__ import print_function, division
import cv2 # opencv
import time
from matplotlib import pyplot as plt # this lets you draw inline pictures in the notebooks
import pylab # this allows you to control figure size 
pylab.rcParams['figure.figsize'] = (10.0, 8.0) # this controls figure size in the notebook
```

# Grab Group Picture

```python
cap = cv2.VideoCapture(0)
ret, frame = cap.read()
if ret:
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame)

cap.release()
```

## Scale and Rotate Image

```python
rot = cv2.transpose(frame)
height, width, colors = frame.shape
scale = cv2.resize(frame,(width//2, height//2), interpolation = cv2.INTER_CUBIC)

plt.subplot(1,2,1)
plt.imshow(rot)
plt.title('rotated')

plt.subplot(1,2,2)
plt.imshow(scale)
plt.title('scaled by 50%');
```

## Add Noise and Filter

# Find Edges

```python
sobelimage=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

sobelx = cv2.Sobel(sobelimage,cv2.CV_64F,1,0,ksize=9)
edges = cv2.Sobel(sobelimage,cv2.CV_64F,0,1,ksize=21)
# edges = abs(edges)
# ret,edges = cv2.threshold(edges,1e11,1e13,cv2.THRESH_BINARY_INV)

# plt.subplot(1,2,1)
# plt.imshow(frame)
# plt.title('original')

# plt.subplot(1,2,2)
plt.imshow(edges,cmap = 'gray') 
plt.colorbar()
plt.title('Edges');
```

# Change Colorspace to HSV

```python
hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

plt.subplot(1,2,1)
plt.imshow(frame)
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(hsv)
plt.title('HSV');
```

# ROI

```python
width, height, colors = frame.shape

plt.imshow(frame[0:width//2, 0:height//2]);
```

# Draw on Image

```python
frame_draw = frame.copy()
height, width, colors = frame.shape
top_left = (width//3, height//3)
bottom_right = (2*width//3, 2*height//3)
print(bottom_right)
cv2.rectangle(frame_draw,top_left,bottom_right,(0,255,0),3)  # 
cv2.circle(frame_draw,(width//2, height//2), 10, (0,0,255), -1)  # -1 means filled
plt.imshow(frame_draw)
```
